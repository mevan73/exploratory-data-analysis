```python
import pandas as pd
df = pd.read_excel("messy-data.xlsx")
df.head()
```

```python
df = pd.read_excel("messy-data.xlsx", skiprows=2)
df.head()
```

```python
df = pd.read_excel("messy-data.xlsx", skiprows=2, usecols="B:D")
df.head(20)
```

```python
df.dropna(inplace=True, axis=0)
df.head(20)
```

```python
splitted_columns = df['Month, period'].str.split(",",expand=True)
splitted_columns
```

```python
df['Month'] = splitted_columns[0]
df['period'] = splitted_columns[1]
df.head(15)
```

```python
df.drop('Month, period', axis=1, inplace=True)
df.head(15)
```

```python
def remove_mm(x):
    if type(x) is str:
        return x.replace('mm', '')
    else:
        return x
```

```python
df['Lake Victoria'] = df['Lake Victoria'].apply(lambda x: remove_mm(x))
df['Simiyu'] = df['Simiyu'].apply(lambda x: remove_mm(x))
df.head(20)
```

```python
df.tail(5)
```

```python
df['Simiyu'] = df['Simiyu'].astype(float)
df
```

```python
def adele1(s):
    if s.Simiyu >= 3.0:
        return['background-color:cyan']*4
    else:
        return['background-color:greenyellow']*4
df.style.apply(adele1,axis=1)
```
