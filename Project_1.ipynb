```python
import numpy as np
import pandas as pd
df3=pd.read_csv(r"C:\Users\User\pyproj\empSalary.csv")
splittedcolumns=df3["Last,First"].str.split(',',expand=True)
df3["FirstName"]=splittedcolumns[0]
df3["lastname"]=splittedcolumns[1]
df3
```

```python
df3.drop(["Last,First"],axis = 1)
```

```python
df3["Price"]=df3["Price"].replace('[^0-9]+','',regex=True)
df3
```

```python
def increase(x):
    if x.Commision>4000:
        return x.Salary+x.Salary*0.2
    else:
        return x.Salary+x.Salary*0.1
df3["NewSalary"]=df3.apply(increase,axis=1)
df3
```

```python
df3.to_csv(r"C:\Users\User\pyproj\empnewsalary.csv")
```

```python
df4.drop(["Last,First"],axis = 1,inplace=True)
```

```python
def bgc(s):
    if s["Product Name"]=="Laptop":
        return["background-color:cyan"]*24
    elif s["Product Name"]=="Camera":
        return["background-color:yellow"]*24
    elif s["Product Name"]=="Mobile":
        return["background-color:blue"]*24
    elif s["Product Name"]=="Computer":
        return["background-color:green"]*24
    else:
        return["background-color:red"]*24
df4.style.apply(bgc,axis=1)
```
