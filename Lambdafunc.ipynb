```python
l=[12,56,89,23,50,14,25,36,78,91,2,3,4,6,5,7,12]
Evenlist=list(filter(lambda x:(x%2==0),l))
print(Evenlist)
```

```python
l=[12,56,89,23,50,14,25,36,78,91,2,3,4,6,5,7,12]
oddlist=list(filter(lambda x:(x%2!=0),l))
print(oddlist)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,3*np.pi,0.2)
y=np.sin(x)
plt.plot(y)
plt.show()
z = np.cos(x)
p = y+z
plt.plot(x,z)
plt.plot(p)
plt.show()
```
