```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-1.5,1.5,30)
px=0.8
py=px**2
plt.plot(x,x**2,"b-",px,py,"ro")
plt.text(0,1.5,"Square function\n$y=x^2$",fontsize=20,color="blue",horizontalalignment="center",weight="heavy")
plt.text(px-0.08,py,"Beautiful point",ha="right",weight="heavy")
plt.text(px,py,"x = %0.2f\ny = %0.2f"%(px,py),rotation=50,color="grey")
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#example 2-subplots
#creating the raw data
category=["Account A","Account B","Account C"]
revenue= np.array([300,450,700])
#plotting it
plt.figure(figsize=(10,5))
# bar
# 1=nrows,3=ncolmn,1=index or position. .==top left
plt.subplot(131)
plt.bar(range(0,3),revenue)
plt.xticks(range(0,3),category)
# line
plt.subplot(132)
plt.plot(range(0,3),revenue, c="red")
plt.xticks(range(0,3),category)
#scatter
plt.subplot(133)#top right
plt.scatter(range(0,3),revenue, c="cyan")
plt.xticks(range(0,3),category)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
fig, ax=plt.subplots()
fruits = ["apple","blueberry","cherry","orange"]
counts = [140,100,30,55]
bar_labels = ["red","blue","cyan","orange"]
bar_colors = ["tab:red","tab:blue","tab:cyan","tab:orange"]
ax.bar(fruits,counts, label=bar_labels, color=bar_colors)
ax.set_ylabel("Fruit Supply")
ax.set_title("Fruit supply by Kind and color")
ax.legend(title="Fruit color")
```
