```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-2,2,500)
y=x**2
plt.plot(x,y)
plt.title("Square function")
plt.xlabel("X",fontsize=14)
plt.ylabel("Y=x**2",fontsize=14)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-2,2,500)
y=-x**2
plt.plot(x,y)
plt.title("Square function")
plt.xlabel("X",fontsize=14)
plt.ylabel("Y=-x**2",fontsize=14)
plt.grid(True)
plt.show()
```
